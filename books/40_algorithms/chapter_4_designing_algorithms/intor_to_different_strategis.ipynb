# DIVIDE AND CONQUER

By example of simplification of apache [spark](https://spark.apache.org/)




```python
words = ["python", "java", "ottawa", "news", "java", "ottawa"]
print(words)
```

```python
import findspark

findspark.init()
```

let's create a sample list containing some words. We will convert this list
into Spark's native distributed data structure, called a Resilient Distributed
Dataset (RDD

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").getOrCreate()
sc = spark.SparkContext

```

Now, let's use a map function to convert the words into a key-value pair:

```python
wordsRDD = sc.parallelize(words, 4)
print(wordsRDD.collect())

word_pairs = wordsRDD.map(lambda w: (w, 1))

print(wordsRDD.collect())
```

```python
word_counts_collect = word_pairs.reduceByKey(lambda x, y: x+y)
print(word_counts_collect.collect())
```
